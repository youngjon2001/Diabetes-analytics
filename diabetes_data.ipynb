# Demographics

### Age Demographics

````python
bins = [0, 29, 39, 49, 59, 69, 150]
labels = ["<30", "30–39", "40–49", "50–59", "60–69", "70+"]

df["age_band"] = pd.cut(
    df["age"],
    bins=bins,
    labels=labels,
    right=True,
    include_lowest=True
)
````

````python
diabetes_counts = (
    df.groupby("age_band")["diagnosed_diabetes"]
      .value_counts()
      .unstack()
)

diabetes_counts[1].plot(kind="bar", stacked=True)

plt.xlabel("age_band")
plt.ylabel("Count")
plt.title("Diabetes Status by Gender")
plt.legend(title="Diabetes Status")
plt.show()
````

### Interpretation of Diagnosed Diabetes Status by Age

This bar plot illustrates the number of individuals diagnosed with diabetes across different age bands. It is evident that the count of diagnosed diabetes cases generally increases with age, with the highest numbers observed in the `50-59` and `60-69` age groups. This trend suggests a higher prevalence or diagnosis rate of diabetes in older populations within this dataset. The `<30` age band shows the lowest number of diagnosed cases, as expected.

### Ethnicity Demographics

````python
df.ethnicity.value_counts()
````

````python

from scipy.stats import chi2_contingency

# 1. Create the 2D contingency table
# This automatically calculates the frequencies for 0 and 1
contingency_table = pd.crosstab(df["ethnicity"], df["diagnosed_diabetes"]) # Corrected column name

# 2. Perform the Chi-Square Test
# Output: stat (test statistic), p (p-value), dof (degrees of freedom), expected (frequencies)
stat, p, dof, expected = chi2_contingency(contingency_table)

# 3. Interpret the result
print(f"P-value: {p}")
if p < 0.05:
    print("Significant difference exists across ethnic groups.")
else:
    print("No significant difference found.")
````

````python


# Create contingency table normalized by row (gives proportions)
ct_prop = pd.crosstab(df['ethnicity'], df["diagnosed_diabetes"], normalize='index')

# Plot
ct_prop.plot(kind='bar', stacked=True, color=['#aec6cf', '#ff6961'])
plt.title('Diabetes Prevalence by Ethnic Group (Proportion)')
plt.ylabel('Proportion (0 to 1.0)')
plt.legend(title='Diabetes', labels=['No (0)', 'Yes (1)'])
plt.show()
````

### Interpretation of Diabetes Prevalence by Ethnic Group

This stacked bar plot visually represents the proportion of individuals diagnosed with diabetes (labeled 'Yes' or '1') versus those not diagnosed (labeled 'No' or '0') within each ethnic group. Consistent with the Chi-Square test's finding of no significant difference (p-value = 0.321), the proportions of diagnosed diabetes are remarkably similar across all ethnic categories. Approximately 60% of individuals in each ethnic group are diagnosed with diabetes, and about 40% are not. This visualization confirms that, in this dataset, ethnicity alone does not appear to be a major differentiating factor in diabetes prevalence.

### Interpretation of the `fam_history:lifestyle_score` Interaction Term

In a logistic regression model, an interaction term allows us to assess whether the effect of one predictor variable on the outcome changes depending on the level of another predictor variable. Specifically, for `fam_history:lifestyle_score`:

*   **`fam_history`**: This is a binary variable indicating whether an individual has a family history of diabetes (1 for 'Yes', 0 for 'No').
*   **`lifestyle_score`**: This is a numerical variable (in this case, `diet_score`) representing a component of lifestyle.

**What the Interaction Term Means:**

If the interaction term `fam_history:lifestyle_score` were statistically significant, it would mean that the effect of `lifestyle_score` on the odds of being diagnosed with diabetes is different for individuals with a family history of diabetes compared to those without. Conversely, it would also imply that the effect of `fam_history` on the odds of diabetes diagnosis changes depending on an individual's `lifestyle_score`.

For example:
*   A **positive significant coefficient** for the interaction term would suggest that a higher `lifestyle_score` (better diet) has a *smaller* protective effect (or even an exacerbating effect) on diabetes risk when there's a family history of diabetes, compared to when there isn't.
*   A **negative significant coefficient** would suggest that a higher `lifestyle_score` (better diet) has a *larger* protective effect on diabetes risk when there's a family history of diabetes, compared to when there isn't.

This finding highlights a crucial target group for early intervention efforts: high-risk individuals under 30 who remain undiagnosed.




### Insights
*   **Targeted Screening for Young Adults:** Prioritize targeted screening and awareness campaigns for high-risk individuals under the age of 30, as they are significantly more likely to be undiagnosed.
*   **Investigate Causes of Underdiagnosis in Younger Groups:** Further research is needed to understand the specific reasons for higher underdiagnosis rates in younger high-risk individuals, such as lower healthcare utilization, less frequent routine check-ups, or lack of awareness among this demographic and healthcare providers.
